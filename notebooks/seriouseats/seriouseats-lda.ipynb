```python
# NLTK Stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
```

```python
def delete_preamble(title):
    if ':' in title:
        title_split = title.split(':')
        return title_split[-1].strip()
    return title

import re
def get_title_post_description(title):
    match = re.search(r".+\((.+)(\)|\.{3})", title)
    if match:
        return match.group(1)
    return title

def remove_the_best_prefix(title):
    match = re.search(r"The Best (.+)", title)
    if match:
        return match.group(1)
    return title

def preprocess(title):
    title = delete_preamble(title)
    title = get_title_post_description(title)
    title = remove_the_best_prefix(title)
    return title

df['title'] = df['title'].map(preprocess)
df.head()
```

```python
titles = df['title'].tolist()
```

```python
import gensim

def titles_to_words(titles):
    for t in titles:
        yield gensim.utils.simple_preprocess(t, deacc=True)
        
data_words = list(titles_to_words(titles))
```

```python
data_words[:10]
```

```python
bigram = gensim.models.Phrases(data_words, min_count=4, threshold=100)
trigram = gensim.models.Phrases(bigram[data_words], threshold=100)
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)
```

```python
print(trigram_mod[bigram_mod[data_words[1]]])
```

```python
import spacy
nlp = spacy.load('en', disable=['parser', 'ner'])
```

```python
def remove_stopwords(texts):
    return [[w for w in gensim.utils.simple_preprocess(str(title)) if w not in stop_words] for title in titles]

def make_bigrams(titles):
    return [bigram_mod[t] for t in titles]

def make_trigrams(titles):
    return [trigram_mod[bigram_mod[t]] for t in titles]

def lemmatization(titles, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    titles_out = []
    for sent in titles:
        doc = nlp(" ".join(sent))
        titles_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return titles_out
```

```python
data_words_nostops = remove_stopwords(data_words)
data_words_bigrams = make_bigrams(data_words_nostops)
data_lemmatized = lemmatization(data_words_bigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])

data_lemmatized[:10]
```

```python
import gensim.corpora as corpora
id2word = corpora.Dictionary(data_lemmatized)
texts = data_lemmatized
corpus = [id2word.doc2bow(text) for text in texts]
corpus[:4]
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=100, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=1000,
                                           passes=20,
                                           alpha='auto',
                                           per_word_topics=True)
```
